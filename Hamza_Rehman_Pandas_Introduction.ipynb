###Filtering Data

There is too much data to take in at once, so now it is your job to filter out specific columns or subsets that you are interested in. To access a specific column of data, ```median_income``` for example, you would call it up like this:

````python
data.median_income
````

Remember that ```data``` here is just the name we gave the variable that holds the housing dataset, this isn't a specific function in the library.

Let's do something practical! What if you were tasked to find the following: all median housing values over $150,000 where the data is sorted according to the median income of the area? This query has two parts:


*   To get that specific subset of housing data, the command would be ```data[data.median_housing_value>150000]```
*   To sort it based on median income, you could add the method ```.sort_values('median_income')```

Together, that would make this query: ```data[data.median_house_value > 150000].sort_values(('median_income')``` Run the next cell to save that query in the variable ```query1```:

````python
query1 = data[data.median_house_value > 150000].sort_values( 'median_income' )
````

You can run ```query1``` to see the results:

````python
query1
````

###Plots and Graphs

Finally, you can visualize the data to help you draw better conclusions and more easily see correlations. For example, to make a scatter plot of the query you just made, run the following:

````python
query1.plot.scatter(x='median_income',y='median_house_value',figsize=(12,12))
````
